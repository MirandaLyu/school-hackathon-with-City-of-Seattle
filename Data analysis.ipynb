```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "2022_Building_Energy_Benchmarking.csv"
data = pd.read_csv(file_path)

data.head(), data.info()
```

```python
data.describe()
```

```python
columns_to_check = [
    'ENERGYSTARScore', 'SiteEUIWN(kBtu/sf)', 'SiteEUI(kBtu/sf)',
    'SiteEnergyUse(kBtu)', 'SiteEnergyUseWN(kBtu)', 'SourceEUIWN(kBtu/sf)',
    'SourceEUI(kBtu/sf)', 'TotalGHGEmissions', 'GHGEmissionsIntensity']

# IQR
outliers_summary = {}

for column in columns_to_check:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_count = data.loc[(data[column] < lower_bound) | (data[column] > upper_bound), column].count()
    outliers_summary[column] = outliers_count

outliers_df = pd.DataFrame.from_dict(outliers_summary, orient='index', columns=['Outliers Count'])
outliers_df
```

```python
# bloxplot
# Set the style of seaborn
sns.set_style("whitegrid")

# Creating boxplots for 'SiteEnergyUse(kBtu)' and 'ENERGYSTARScore'
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Site Energy Use (kBtu)
sns.boxplot(data=data, y='SiteEnergyUse(kBtu)', ax=ax[0])
ax[0].set_title('Site Energy Use (kBtu)')
ax[0].set_ylabel('Energy Use (kBtu)')

# ENERGYSTARScore
# sns.boxplot(data=data, y='ENERGYSTARScore', ax=ax[1])
# ax[1].set_title('ENERGYSTARScore')
# ax[1].set_ylabel('Score')

# GHGEmissionsIntensity
sns.boxplot(data=data, y='GHGEmissionsIntensity', ax=ax[1])
ax[1].set_title('GHG Emissions Intensity')
ax[1].set_ylabel('Intensity')

plt.tight_layout()
plt.show()
```

```python
# compliance status distribution 'EPAPropertyType' and 'ComplianceStatus'
compliance_status_grouped = data.groupby(['EPAPropertyType', 'ComplianceStatus']).size().unstack(fill_value=0)
compliance_status_grouped
```

```python
compliance_status_percentage = compliance_status_grouped.apply(lambda x: (x / x.sum()) * 100, axis=1)

compliance_status_percentage_sorted = compliance_status_percentage.sort_values(by='Not Compliant', ascending=False)
compliance_status_percentage_sorted
```

```python
# Calculate the frequency of each category in 'ComplianceIssue' column
compliance_issue_freq = data['ComplianceIssue'].value_counts()

# Display the frequency distribution
compliance_issue_freq
```

```python
# Anomaly Detection for 'SiteEnergyUse(kBtu)' using Z-score to identify outliers
from scipy.stats import zscore

# Compute Z-scores for 'SiteEnergyUse(kBtu)'
data['SiteEnergyUse_zscore'] = zscore(data['SiteEnergyUse(kBtu)'].dropna())

# Identify outliers as those with a Z-score greater than 3 or less than -3 (typical threshold for outliers)
outliers_energy_use = data.loc[data['SiteEnergyUse_zscore'].abs() > 3]

# Consistency Check for Gross Floor Area
# Checking if the sum of 'PropertyGFABuilding(s)' and 'PropertyGFAParking' equals 'PropertyGFATotal'
data['GFA_Check'] = (data['PropertyGFABuilding(s)'] + data['PropertyGFAParking']) == data['PropertyGFATotal']

# Summary of findings
outliers_summary = {
    'Outliers in SiteEnergyUse(kBtu)': outliers_energy_use.shape[0],
    'Inconsistencies in GFA Reporting': (~data['GFA_Check']).sum()
}

print(outliers_summary)

# Display the rows corresponding to the 30 outliers identified in 'SiteEnergyUse(kBtu)'
outliers_energy_use[['OSEBuildingID', 'BuildingName', 'SiteEnergyUse(kBtu)']]

#SWEDISH MEDICAL CENTER CHERRY HILL - CAMPUS (OSEBuildingID: 268) is listed as "Not Compliant."
```
